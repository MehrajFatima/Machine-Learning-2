```python
import pandas as pd
```

```python
ds = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Ice%20Cream.csv')
```

```python
y = ds['Revenue']
x = ds[['Temperature']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 2529)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pre = model.predict(x_test)
y_pre
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.metrics import mean_absolute_percentage_error
res = mean_absolute_percentage_error(y_test,y_pre)
```

```python
print("Accuracy = ",100-res*100)
```
